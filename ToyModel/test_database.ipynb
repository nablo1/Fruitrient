```python
import pandas as pd
import sqlite3
from sqlite3 import Error

#Create pandas df from csv train file
df = pd.read_csv('input/train.csv')

#Connect to DB
def db_connection(db_name):
    conn = None
    try:
        conn = sqlite3.connect(db_name)
        return conn
    except Error as e:
        print(e)
        
    return conn

db_conn = db_connection('test_db')
    
#Import data from train dataframe to database
df.to_sql(name='fruits', con=db_conn, if_exists='append')
```
